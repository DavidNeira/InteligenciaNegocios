```python
import pandas as pd
df = pd.read_csv('datos_final.csv',encoding='latin1')
df.head()
```

```python
rowto_string = ['entCode','CodigoProductoONU']
rowto_num = ['CantidadItem', 'CantidadOferta']
df[rowto_string] = df[rowto_string].astype('string')
df[rowto_num] = df[rowto_num].astype('int64')
df.info()
```

```python
pd.set_option('display.float_format', lambda x: '%.f' % x)

df.describe()
```

```python
df.describe(include='object')
```

```python
import matplotlib.pyplot as plt

# Calcular el conteo de ocurrencias de cada tamaño de proveedor
conteo_tamano_proveedor = df['TamanoProveedor'].value_counts()

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
conteo_tamano_proveedor.plot(kind='bar')
plt.title('Número de ocurrencias por tamaño de proveedor')
plt.xlabel('Tamaño del Proveedor')
plt.ylabel('Número de Ocurrencias')
plt.xticks(rotation=45)
plt.show()

# Mostrar la tabla de conteo
print("Tabla de conteo por tamaño de proveedor:")
print(conteo_tamano_proveedor)
```
