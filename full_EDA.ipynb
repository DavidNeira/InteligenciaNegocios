```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Función para remover outliers basados en el método IQR y graficar el boxplot
def remover_outliers_y_graficar(df, iteracion):
    # Calcule el primer y tercer cuartil
    Q1 = df['MontoNetoOC'].quantile(0.25)
    Q3 = df['MontoNetoOC'].quantile(0.75)

    # Calcule el rango intercuartílico (IQR)
    IQR = Q3 - Q1

    # Defina el límite inferior y superior para outliers
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Identifique los outliers
    outliers = df[(df['MontoNetoOC'] < limite_inferior) | (df['MontoNetoOC'] > limite_superior)]
    df_limpiado = df.drop(outliers.index)

    # Graficar el boxplot después de la iteración
    plt.figure()
    sns.boxplot(x='MontoNetoOC', data=df_limpiado)
    plt.ticklabel_format(style='plain', axis='x')


    # Agrega líneas verticales para los cuartiles
    plt.axvline(Q1, color='r', linestyle='--', label='Cuartil 1 (Q1)')
    plt.axvline(Q3, color='g', linestyle='--', label='Cuartil 3 (Q3)')

    # Muestra los valores de los cuartiles en el gráfico
# Muestra los valores de los cuartiles en el gráfico
    plt.text(Q1, 0, f'Q1 = {Q1:.2f}', rotation=0, va='top', ha='right', color='r', bbox=dict(facecolor='white', alpha=0.5))
    plt.text(Q3, 0, f'Q3 = {Q3:.2f}', rotation=0, va='top', ha='left', color='g', bbox=dict(facecolor='white', alpha=0.5))


    # Muestra los cuartiles en la leyenda
    plt.legend()

    # Muestra información sobre la iteración y los outliers
    plt.title(f"Iteración {iteracion}: Nuevo rango [{df_limpiado['MontoNetoOC'].min()}, {df_limpiado['MontoNetoOC'].max()}], Outliers: {len(outliers)}")
    plt.show()

    return df_limpiado, outliers, df_limpiado

# Cargar el DataFrame
df = pd.read_csv("./dataset/licitacion.csv", sep=';', encoding='latin1', decimal=',')

# Iterar hasta que no haya más outliers
iteracion = 1
while True:
    df, outliers, df_actualizado = remover_outliers_y_graficar(df, iteracion)
    nuevo_rango_min = df_actualizado['MontoNetoOC'].min()
    nuevo_rango_max = df_actualizado['MontoNetoOC'].max()
    print(f"Iteración {iteracion}: Nuevo rango [{nuevo_rango_min}, {nuevo_rango_max}]")
    if outliers.empty:
        print("No se encontraron outliers en esta iteración.")
        break
    iteracion += 1
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Función para remover outliers basados en el método IQR y graficar el boxplot
def remover_outliers_y_graficar(df, variable, iteracion):
    # Calcule el primer y tercer cuartil
    Q1 = df[variable].quantile(0.25)
    Q3 = df[variable].quantile(0.75)

    # Calcule el rango intercuartílico (IQR)
    IQR = Q3 - Q1

    # Defina el límite inferior y superior para outliers
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Identifique los outliers
    outliers = df[(df[variable] < limite_inferior) | (df[variable] > limite_superior)]
    df_limpiado = df.drop(outliers.index)

    # Graficar el boxplot después de la iteración
    plt.figure()
    sns.boxplot(x=variable, data=df_limpiado)

    plt.title(f"Iteración {iteracion}: Nuevo rango [{df_limpiado[variable].min()}, {df_limpiado[variable].max()}], Outliers: {len(outliers)}")
    plt.show()

    return df_limpiado, outliers, df_limpiado

# Cargar el DataFrame
df = pd.read_csv("./dataset/licitacion.csv", sep=';', encoding='latin1', decimal=',')

# Lista de variables cuantitativas
cuantitativas = [ 'MontoNetoOC', 'ImpuestosOC', 'DescuentosOC', 'MontoTotalOC', 
                 'ImpuestosOC_CLP', 'MontoNetoOC_CLP', 'CantidadItem', 'MontoNetoItem', 
                 'DescuentoItem', 'CargosItem', 'ImpuestoEspecificoItem', 'MontoTotalItem', 
                 'MontoNetoItemCLP']

# Iterar hasta que no haya más outliers para cada variable cuantitativa
for variable in cuantitativas:
    iteracion = 1
    while True:
        df, outliers, df_actualizado = remover_outliers_y_graficar(df, variable, iteracion)
        nuevo_rango_min = df_actualizado[variable].min()
        nuevo_rango_max = df_actualizado[variable].max()
        print(f"Iteración {iteracion}: Nuevo rango [{nuevo_rango_min}, {nuevo_rango_max}]")
        if outliers.empty:
            print("No se encontraron outliers en esta iteración.")
            break
        iteracion += 1
```
