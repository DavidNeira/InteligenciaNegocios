```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

#Reading the CSV file 
df = pd.read_csv("limpio.csv") 
  
# Printing top 5 rows 
df.head()
```

```python
df.describe()
```

```python
df.head()
```

```python
describe_result = df.describe(include=['object'])

# Contar valores nulos por columna
nulos = df.isnull().sum()

# Agregar la fila 'nulos' al resultado de describe()
describe_result.loc['nulos'] = nulos

# Calcular porcentaje de valores nulos por columna
num_rows = len(df)
null_percentages = (nulos / num_rows) * 100

# Formatear los porcentajes con el símbolo de porcentaje
null_percentages_formatted = null_percentages.apply(lambda x: f"{x:.2f}%")

# Agregar la fila 'porcentaje de nulos' al resultado de describe()
describe_result.loc['porcentaje de nulos'] = null_percentages_formatted
describe_result
```

```python
df.columns
```

```python
# Eliminar múltiples columnas del DataFrame
df_new = df.drop(columns=['NroLicitacion','FuenteFinanciamiento', 'JustificacionMontoEstimado'])
```

```python
# Filtrar licitaciones públicas
licitaciones_publicas = df_new[df_new['TipoLicitacion'].str.contains('Pública')]

# Filtrar licitaciones privadas
licitaciones_privadas = df_new[df_new['TipoLicitacion'].str.contains('Privada')]

# Definir colores para cada tipo de licitación
colores = ['#1f77b4', '#ff7f0e']  # Colores para público y privado respectivamente

# Crear subgráficos para cada tipo de licitación
plt.figure(figsize=(10, 5))

# Gráfico para licitaciones públicas
plt.subplot(1, 2, 1)
sns.countplot(y='TipoLicitacion', data=licitaciones_publicas, palette=[colores[0]])

# Etiquetas y título
plt.title('Licitaciones Públicas')

# Eliminar etiquetas del eje y
plt.ylabel('')

# Gráfico para licitaciones privadas
plt.subplot(1, 2, 2)
sns.countplot(y='TipoLicitacion', data=licitaciones_privadas, palette=[colores[1]])

# Etiquetas y título
plt.title('Licitaciones Privadas')

# Eliminar etiquetas del eje y
plt.ylabel('')

# Ajustar diseño
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```
